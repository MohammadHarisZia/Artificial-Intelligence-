<h1>Customer Churn Analysis<h1>

```python
# Importing the libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
# Importing the dataset
df = pd.read_csv('Customer-Churn.csv')
```

```python
df.head()
```

```python
#Checking types of all the columns to identify which ones to Encode into numerics
df.dtypes
```

<h3>Data Encoding can be done using two Methods of the Sklearn Library, One Hot Encoding and Label Encoding. We will be using the Label Encoder class to transform our data for the model to train</h3>

```python
number = LabelEncoder()

new_array=['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod','TotalCharges','Churn']

for i in new_array:
    df[i] = number.fit_transform(df[i].astype(str))
    
```

```python
# Checking missing values
df.isna().sum()
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
#Dropping the CustomerID Column as it adds zero value to the Model Training 
df = df.drop("customerID",axis=1)
df.dtypes
```

```python
#Dropping the Churn Column as it is our dependant variable by updating our input variable X and storing it in the output variable y.
X=df.drop("Churn",axis=1)
y=df["Churn"]
```

```python
df.head(10)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
df.dtypes
```

```python
model = GaussianNB()

model.fit(X_train,np.ravel(y_train))

y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
```
