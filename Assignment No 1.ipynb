```python
class Node:
    def __init__ (self,state,parent,actions,heuristic,totalCost):
        self.state=state
        self.parent=parent
        self.actions=actions
        self.heuristic=heuristic
        self.totalCost=totalCost
```

```python
def actionSequence(graph,initialState,goal):
    solution=[goal]
    currP=graph[goal].parent
    while currP != None:
        solution.append(currP)
        currP=graph[currP].parent
    solution.reverse()
    return solution
```

```python
import math
def findMin(frontier):
    minV=math.inf
    node=''
    for i in frontier:
        if minV>frontier[i][1]:
            minV=frontier[i][1]
            node=i
    return node
```
